## Generating the data

```python
import SyntheticGenerator as sg
# The path for the data to be written in
data_write_path = "./data/"

# List of TS sequence length to be used
sequence_lengths = [10, 50]

# How many samples to be generated for each class
num_samples_per_class = 10

# How many files/times to run this process. each time generates: num_samples_per_class * num_classes (default is using)
# all classes which is 13. 
num_times = 2

# generate data
sg.generate_synthetic_data(data_write_path, sequence_lengths, num_samples_per_class, num_times)
```

## Let's look at the generated data

```python
from compress_pickle import load
x = load("./data/50/10/50_part0_x_test.gz")
```

## Let's plot some of the Time series generated

```python
import matplotlib.pyplot as plt
import sequences
```

```python
## print all 13 classes that was generated by us
classes = sequences.get_all_sequences()
print(classes)
```

## No noise added to the TS generated data

```python
idxs = [i for i in range(1,x.shape[0]-1, num_samples_per_class)]
for idx, sample in zip(idxs, classes):
    plt.plot(x[idx])
    plt.title(sample)
    plt.show()
```

## with noise added (chaning std on kwargs["other"] on SyntheticGenerator.py to 0.2 from 0)

```python
idxs = [i for i in range(1,x.shape[0]-1, num_samples_per_class)]
for idx, sample in zip(idxs, classes):
    plt.plot(x[idx])
    plt.title(sample)
    plt.show()
```
